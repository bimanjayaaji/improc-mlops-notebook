```python
# importing

import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# cv basics course 1 (3.12)

```python
img = cv2.imread('data/00-puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# important

# matplotlib --> RGB

# opencv --> BGR
```

```python
fix_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_image)
```

```python
img_gray = cv2.imread('data/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
img_resized = cv2.resize(fix_image,(1000,400))
```

```python
plt.imshow(img_resized)
```

```python
img_resized.shape
```

```python
w_ratio = 0.5
h_ratio = 0.5
img_resized2 = cv2.resize(fix_image,(0,0),fix_image,w_ratio,h_ratio)
```

```python
plt.imshow(img_resized2)
```

```python
img_resized2.shape
```

```python
img_resized2.size
```

```python
img_flipped = cv2.flip(fix_image,0)
```

```python
plt.imshow(img_flipped)
```

```python
img_flipped = cv2.flip(fix_image,1)
plt.imshow(img_flipped)
```

```python
img_flipped = cv2.flip(fix_image,-1)
plt.imshow(img_flipped)
```
