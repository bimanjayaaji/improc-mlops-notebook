# cv basic course 4 (3.15)

```python
# add text

blank_img2 = np.zeros(shape=(512,512,3),dtype=np.int16)

font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img2,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(blank_img2)
```

```python
# custome polygon

blank_img3 = np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
plt.imshow(blank_img3)
```

```python
vertices = np.array([ [100,300], [200,450], [300,500], [400,100] ],dtype=np.int32)
```

```python
vertices
```

```python
vertices.shape
```

```python
# need to reshape

```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts
```

```python
# could just make it 3 dimensions in the first place
```

```python
cv2.polylines(blank_img3,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img3)
```

# cv basic course 5 (3.16)

```python
# direct drawing with mouse
```

```python
# run this in a py script

import cv2
import numpy as np

def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,255,0),-1)

cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing',draw_circle)

img = np.zeros((512,512,3),np.int8) # will make the image looks gray

while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

# cv basic course 6 (3.17)

```python
# direct drawing with mouse pt 2
```

```python
# run this in a py script

import cv2
import numpy as np

def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,255,0),-1)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),40,(0,255,0),5)

cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing',draw_circle)

img = np.zeros((512,512,3),np.int8) # will make the image looks gray

while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

# cv basic course 7 (3.18)

```python
# direct drawing with mouse pt 3
```

```python
# run this in a py script

import cv2
import numpy a np

drawing = False
ix,iy = -1,-1

def draw_rect(event,x,y,flags,params):
    global ix,iy,drawing
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)

img = np.zeros((512,512,3))

cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing')

while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break;

cv2.destroyAllWindows()
    
```
