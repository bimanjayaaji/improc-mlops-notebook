```python
res = cv2.matchTemplate(full,face,eval('cv2.TM_CCOEFF'))
```

```python
plt.imshow(res)
```

```python
# A HEAT MAP. COOL!
```

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    
    # create a copy 
    full_copy = full.copy()

    method = eval(m)

    # template matching
    res = cv2.matchTemplate(full_copy,face,method)

    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)

    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]: # these methods are working on a min value
        tl = min_loc
    else:
        tl = max_loc

    height,width,channels = face.shape

    br = (tl[0]+width,tl[1]+height)

    cv2.rectangle(full_copy,tl,br,(255,0,0),10)

    #plot
    plt.subplot(121)
    plt.imshow(res)
    plt.title('heatmap')
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('template matching')

    plt.suptitle(m)

    plt.show() # to show all the plot and not overriding each other

    print('\n')
```
